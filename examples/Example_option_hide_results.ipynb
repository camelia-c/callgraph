```python
from functools import lru_cache
import callgraph
```

```python
%load_ext callgraph
```

```python
@lru_cache()
def Fibonacci(n):
    if n == 0: 
        return 0
    elif n == 1: 
        return 1
    else: 
        return Fibonacci(n-1) + Fibonacci(n-2)
```

```python
def Fibonacci_sequence(n):
    for i in range(n+1):
        print(i,":", Fibonacci(i))
```

```python
Fibonacci_sequence(n = 10)
```

<b>With default options, the result is written in the nodes</b>

```python
%callgraph Fibonacci(10)
```

<b>With the option <i> -r </i> the result is written on the edge directed from the called function to its calling function</b>

```python
%callgraph -r Fibonacci(10)
```

<b>Now with the new option <i>--hide-results</i> the results are not displayed any more</b>

```python
%callgraph --hide-results Fibonacci(10)
```

<b> With the new option <i> --hide-results </i> combined with the <i> -r </i> option </b>

```python
%callgraph --hide-results -r Fibonacci(10)
```
